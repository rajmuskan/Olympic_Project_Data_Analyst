```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load Dataset
athletes = pd.read_csv('olympic.csv')
```

```python
#check null values
nan_values = athletes.isna()
nan_columns = nan_values.any()
nan_columns
```

```python
athletes.isnull().sum()
```

```python
athletes["Age"].fillna((athletes["Age"].mean()), inplace = True)
athletes["Height (cm)"].fillna((athletes["Height (cm)"].mean()), inplace = True)
athletes["Weight (kg)"].fillna((athletes["Weight (kg)"].mean()), inplace = True)
```

```python
athletes.head()
```

```python
athletes["Medal"].fillna('No Medal', inplace = True)
```

```python
athletes.head()
```

```python
athletes.isnull().sum()
```

```python
athletes.rename(columns={'Team':'Country'}, inplace=True)
```

```python
athletes.head()
```

```python
athletes.to_csv("olympic_modified.csv")
```
